## The First Four Moments: Mean, Variance, Skew, and Kurtosis

When we talk about a probablity dist, we often care about at four things:
- The expected value, or where a distribution is centered (mean)
- The amount of variation from that centered value (variation)
- Whether or not our distribution favors, leans towards, or skews towards certain values. (skew)
- How often to expect unusually high or low values. (kurtosis)

**Task10**:

Describe what process each of the previous discrete distributions models, and their support.

Hint: Make sure to read any of the preliZ documentation, the wikipedia articles, or wolfram alpha.


Bernoulli: single model of binomial distribution. Outcomes can be 0 or 1

Binomial: multiple independent bernoulli trials. Number of successes can be 0 to n trials

Poisson: Number of events occuring in fixed interval time, assuming constant average rate of events. Support is any non-negative integer

Geomteric: Models the number of independent Bernoulli trials it takes to land the first "1" value. Support is any integer including and above 1.

HyperGeometric: Number of successes in n draws from population size N, where there are K successes in the population. Support is between 0 and min(N,k)

Categorical: Models a single trial with multiple outcomes, each with a certain probability. Support is typically given as integers up to the number of possible outcomes.
